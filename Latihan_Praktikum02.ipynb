```python
import pandas as pd
df = pd.read_csv(path + '/Data/day.csv')
df
```

```python
df.info()
```

```python
# Menghitung Nilai-Nilai Sentral (Mean, Median, Modus)
print("=== Nilai Rata-Rata (Mean) ===")
print(df.mean(numeric_only=True))

print("\n=== Nilai Tengah (Median) ===")
print(df.median(numeric_only=True))

print("\n=== Nilai yang Sering Muncul (Modus) ===")
print(df.mode().iloc[0])
```

```python
# Menghitung Ukuran Persebaran (Variansi & Standar Deviasi)
print("=== Variansi ===")
print(df.var(numeric_only=True))

print("\n=== Standar Deviasi ===")
print(df.std(numeric_only=True))
```

```python
# Menghitung Kuartil
print("=== Kuartil Pertama (Q1), Kuartil Ketiga (Q3), dan IQR ===")
for col in df.select_dtypes(include='number').columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    print(f"{col}: Q1={Q1}, Q3={Q3}, IQR={IQR}")
```

```python
# Menghitung Statistik Deskriptif Otomatis
print("=== Statistik Deskriptif Otomatis ===")
df.describe()
```

```python
# Boxplot
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
df.boxplot()
plt.suptitle('Boxplot Semua Variabel Numerik Dataset day.csv')
plt.show()
```

```python
# Histogram
df.hist(figsize=(12,8), bins=10)
plt.suptitle('Histogram Distribusi Data day.csv')
plt.show()
```

```python
# Scatter Plot (contoh hubungan dua variabel numerik)
# Misalnya kita gunakan 'temp' dan 'cnt' dari dataset day.csv
plt.figure(figsize=(8,5))
plt.scatter(df['temp'], df['cnt'])
plt.xlabel('Temperature')
plt.ylabel('Count (Jumlah Penyewaan)')
plt.title('Scatter Plot: Hubungan Temperature vs Count')
plt.show()
```

```python
# Membagi dataset day.csv menjadi tiga bagian:
# (a) Data Training: 80% dari total dataset
# (b) Data Validation: 10% dari data training
# (c) Data Testing: 20% dari total dataset

from sklearn.model_selection import train_test_split

# Langkah 1: Membagi dataset menjadi Data Training (80%) dan Testing (20%)
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Langkah 2: Dari Data Training, ambil 10% untuk Validation
train_data_final, val_data = train_test_split(train_data, test_size=0.1, random_state=42)
```

```python
# Menampilkan jumlah data untuk masing-masing set
print("=== Jumlah Data Tiap Set ===")
print("Data Training :", len(train_data_final))
print("Data Validation :", len(val_data))
print("Data Testing :", len(test_data))
```

```python
# Menampilkan 5 baris data teratas untuk setiap set
print("\n=== 5 Baris Data Teratas - Training Set ===")
display(train_data_final.head())

print("\n=== 5 Baris Data Teratas - Validation Set ===")
display(val_data.head())

print("\n=== 5 Baris Data Teratas - Testing Set ===")
display(test_data.head())
```
